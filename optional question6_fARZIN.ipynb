The following Graph shows that as the imdb_score increases the gross value of the movies would generally increase

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
df2=pd.read_csv('imdb.csv',escapechar='\\')
df4=pd.read_csv('movie_metadata.csv')
```

```python
sns.lmplot(x='imdb_score',y='gross',data=df4)
```

no conclusive result can be extracted from relation between budget and gross

```python
sns.jointplot(x='budget',y='imdb_score',data=df4,kind='scatter')
```

```python
sns.jointplot(x='title_year',y='budget',data=df4,kind='scatter')
```
